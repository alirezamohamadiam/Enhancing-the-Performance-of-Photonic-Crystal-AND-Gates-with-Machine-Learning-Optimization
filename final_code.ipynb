```python
# Import necessary libraries
import pandas as pd
from lazypredict.Supervised import LazyRegressor
from sklearn.utils import shuffle
import numpy as np

# Assuming you have loaded your data1 and data3 DataFrames

# Create empty lists to store X and Y sets
X_sets = []
Y_sets = []

# For data1
X_R11 = data1.iloc[:, 0].values
Y_R11 = data1.iloc[:, 1].values
X_R10 = data1.iloc[:, 0].values
Y_R10 = data1.iloc[:, 2].values

# Append X and Y sets to the lists
X_sets.append(X_R11)
Y_sets.append(Y_R11)

X_sets.append(X_R10)
Y_sets.append(Y_R10)

# For data3
X_X11 = data3.iloc[:, 0].values
Y_X11 = data3.iloc[:, 1].values
X_X10 = data3.iloc[:, 0].values
Y_X10 = data3.iloc[:, 2].values

# Append X and Y sets to the lists
X_sets.append(X_X11)
Y_sets.append(Y_X11)

X_sets.append(X_X10)
Y_sets.append(Y_X10)

# Create LazyRegressor models for each set
models = []

for i in range(4):
    X = X_sets[i]
    Y = Y_sets[i]

    # Shuffle the data
    X, Y = shuffle(X, Y, random_state=13)

    # Split the data into training and testing sets (90% for training and 10% for testing)
    offset = int(X.shape[0] * 0.9)
    X_train, Y_train = X[:offset], Y[:offset]
    X_test, Y_test = X[offset:], Y[offset:]

    reg = LazyRegressor(verbose=0, ignore_warnings=False, custom_metric=None)
    models_data = reg.fit(X_train, X_test, Y_train, Y_test)

    models.append(models_data)

# Print models for each set
for i in range(4):
    print(f"Models for Set {i + 1}:")
    print(models[i])
```
